```python
# Importació de llibreries
import medmnist
from medmnist import INFO
from torchvision import transforms
from torch.utils.data import DataLoader
import matplotlib.pyplot as plt
import numpy as np
```

#DATASETS DISPONIBLES

```python
for name in INFO.keys():
    print("-", name)
```

PATHMIST

```python
# ⚙️ Carga del dataset

# Selecciona el subset que quieres explorar (puedes cambiarlo a 'chestmnist' o 'organamnist')
dataset_name = "pathmnist"
info = INFO[dataset_name]

# Obtén la clase correspondiente del dataset
DataClass = getattr(medmnist, info['python_class'])

# Define las transformaciones (tensor + normalización)
transform = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize(mean=[.5], std=[.5])
])

# Carga el dataset (modo entrenamiento)
train_dataset = DataClass(split='train', transform=transform, download=True)
```

```python
# 🖼️ Visualización de imágenes

fig, axs = plt.subplots(3, 3, figsize=(6, 6))
for i in range(9):
    img, label = train_dataset[i]
    axs[i // 3, i % 3].imshow(img.permute(1, 2, 0))  # Convertir de (C, H, W) a (H, W, C)
    axs[i // 3, i % 3].set_title(f"Label: {label}")
    axs[i // 3, i % 3].axis('off')

plt.tight_layout()
plt.show()
```
